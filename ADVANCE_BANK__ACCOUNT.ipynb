```python
# Simple login system and bank operations

# Set your password
password = "mySecret123"

# Login phase
entered_password = input("Enter your password to login: ")
if entered_password == password:
    print("Login successful!")
else:
    print("Wrong password!")
    exit()  # Exit if password is wrong

# Account setup
print("WELCOME TO WMAXLINE NOVA BANK")
account_number = input("Enter your account number: ")
account_password = input("Enter your account password: ")
account_holder = input("Enter your account holder name: ")

# Check account password (same as login password here)
if account_password == password:
    print(f"WELCOME TO YOUR ACCOUNT, {account_holder.upper()}")
else:
    print("INVALID PASSWORD")
    exit()

# Account class definition
class Account:
    def __init__(self, account_number, account_holder, account_balance=0):
        self.account_number = account_number
        self.account_holder = account_holder
        self.account_balance = account_balance
        self.transaction_history = []

    def deposit(self, amount):
        if amount > 0:
            self.account_balance += amount
            self.transaction_history.append(f"Deposited: {amount}")
            print(f"Deposited {amount} successfully.")
        else:
            print("INVALID, THE DEPOSIT SHOULD BE GREATER THAN 0")

    def withdraw(self, amount):
        if amount <= 0:
            print("INVALID, THE AMOUNT SHOULD BE GREATER THAN 0")
        elif amount > self.account_balance:
            print("INSUFFICIENT FUNDS IN YOUR ACCOUNT")
        else:
            self.account_balance -= amount
            self.transaction_history.append(f"Withdrew: {amount}")
            print(f"Withdrawal of {amount} successful.")

    def check_balance(self):
        return self.account_balance

    def show_transactions(self):
        if self.transaction_history:
            print("Transaction History:")
            for tx in self.transaction_history:
                print("-", tx)
        else:
            print("No transactions yet.")

# Create the account
my_account = Account(account_number, account_holder)

# Menu loop
while True:
    print("\n1. Deposit\n2. Withdraw\n3. Check Balance\n4. Transaction History\n5. Exit")
    choice = input("Choose an option: ")

    if choice == "1":
        try:
            amount = float(input("Enter amount to deposit: "))
            my_account.deposit(amount)
        except ValueError:
            print("Please enter a valid number.")
    elif choice == "2":
        try:
            amount = float(input("Enter amount to withdraw: "))
            my_account.withdraw(amount)
        except ValueError:
            print("Please enter a valid number.")
    elif choice == "3":
        print("Current Balance:", my_account.check_balance())
    elif choice == "4":
        my_account.show_transactions()
    elif choice == "5":
        print("Thank you for banking with WMAXLINE NOVA BANK!")
        break
    else:
        print("Invalid option. Try again.")
```
