```python
# groupfinancial loss per year from cybersecurity threats
result = df.groupby('Year')['Financial Loss (in Million $)'].sum()
# plot a line chart showing annnual financial loss
result.plot(kind='line',title='Annuala financial losss in million dollars $ from threats',ylabel='Financial Loss (in Million $)',marker = 'o')
```
