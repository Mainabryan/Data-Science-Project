```python
# import sys
# # Password check
# password = "Mainabryan1"
# entered_password = input("Enter your password: ")

# if entered_password == password:
    # print("Access granted.")
import sys
# Password check
password = "Mainabryan1"
entered_password = input("Enter your password: ")

if entered_password == password:
    print("Access granted.")
else:
    print("Incorrect password. Access denied.")
    sys.exit()

# Welcome message and user input
print("Welcome to Ereto Library")
library_No = input("Enter your library number: ")
borrowed_books = int(input("Enter the borrowed books: "))
returned_books = int(input("Enter the returned books: "))
members_name = input("Enter the member's name: ")

# Library class
class Library:
    def __init__(self, library_No, borrowed_books, returned_books, members_name):
        self.library_No = library_No
        self.borrowed_books = borrowed_books
        self.returned_books = returned_books
        self.members_name = members_name
        self.personal_data = []

    def borrow_books(self, books):
        if books > 0:
            self.borrowed_books += books
            self.personal_data.append(f"Borrowed {books}")
            print(f"Borrowed {books} successfully.")
        else:
            print("Invalid number of books to borrow.")

    def return_books(self, books):
        if books > 0:
            self.returned_books += books
            self.personal_data.append(f"Returned {books}")
            print(f"Returned {books} successfully.")
        else:
            print("Invalid number of books to return.")

    def display_personal_data(self):
        print("\n-- Personal Data --")
        print(f"Library Number: {self.library_No}")
        print(f"Borrowed Books: {self.borrowed_books}")
        print(f"Returned Books: {self.returned_books}")
        print(f"Member's Name: {self.members_name}")
        print("Activity Log:")
        for data in self.personal_data:
            print(f"- {data}")

# Create library account object
my_library_account = Library(library_No, borrowed_books, returned_books, members_name)

# Main menu loop
while True:
    print("\nLibrary Management System")
    print("1. Borrow Books")
    print("2. Return Books")
    print("3. Display Personal Data")
    print("4. Exit")

    choice = input("Enter your choice (1/2/3/4): ")

    if choice == "1":
        try:
            num_books = int(input("Enter the number of books to borrow: "))
            my_library_account.borrow_books(num_books)
        except ValueError:
            print("Please enter a valid number.")
    elif choice == "2":
        try:
            num_books = int(input("Enter the number of books to return: "))
            my_library_account.return_books(num_books)
        except ValueError:
            print("Please enter a valid number.")
    elif choice == "3":
        my_library_account.display_personal_data()
    elif choice == "4":
        print("Thank you for using the Library Management System. Goodbye!")
        break
    else:
        print("Invalid choice. Please try again.")
# print("Welcome to Ereto Library")
# library_No = input("Enter your library number: ")
# borrowed_books = int(input("Enter the borrowed books: "))
# returned_books = int(input("Enter the returned books: "))
# members_name = input("Enter the member's name: ")

# # Library class
# class Library:
#     def __init__(self, library_No, borrowed_books, returned_books, members_name):
#         self.library_No = library_No
#         self.borrowed_books = borrowed_books
#         self.returned_books = returned_books
#         self.members_name = members_name
#         self.personal_data = []

#     def borrow_books(self, books):
#         if books > 0:
#             self.borrowed_books += books
#             self.personal_data.append(f"Borrowed {books}")
#             print(f"Borrowed {books} successfully.")
#         else:
#             print("Invalid number of books to borrow.")

#     def return_books(self, books):
#         if books > 0:
#             self.returned_books += books
#             self.personal_data.append(f"Returned {books}")
#             print(f"Returned {books} successfully.")
#         else:
#             print("Invalid number of books to return.")

#     def display_personal_data(self):
#         print("\n-- Personal Data --")
#         print(f"Library Number: {self.library_No}")
#         print(f"Borrowed Books: {self.borrowed_books}")
#         print(f"Returned Books: {self.returned_books}")
#         print(f"Member's Name: {self.members_name}")
#         print("Activity Log:")
#         for data in self.personal_data:
#             print(f"- {data}")

# Create library account object
# my_library_account = Library(library_No, borrowed_books, returned_books, members_name)

# # Main menu loop
# while True:
#     print("\nLibrary Management System")
#     print("1. Borrow Books")
#     print("2. Return Books")
#     print("3. Display Personal Data")
#     print("4. Exit")

#     choice = input("Enter your choice (1/2/3/4): ")

#     if choice == "1":
#         try:
#             num_books = int(input("Enter the number of books to borrow: "))
#             my_library_account.borrow_books(num_books)
#         except ValueError:
#             print("Please enter a valid number.")
#     elif choice == "2":
#         try:
#             num_books = int(input("Enter the number of books to return: "))
#             my_library_account.return_books(num_books)
#         except ValueError:
#             print("Please enter a valid number.")
#     elif choice == "3":
#         my_library_account.display_personal_data()
#     elif choice == "4":
#         print("Thank you for using the Library Management System. Goodbye!")
#         break
#     else:
#         print("Invalid choice. Please try again.")
```
