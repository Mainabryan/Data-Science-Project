```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data (corrected)
house_sizes = np.array([50, 75, 125, 150]).reshape(-1, 1)
house_prices = np.array([15000, 20000, 25000, 30000])

# Model training
model = LinearRegression()
model.fit(house_sizes, house_prices)

# Prediction
predicted_prices = model.predict(house_sizes)

# Plotting
plt.scatter(house_sizes, house_prices, color='blue', label='Actual Prices')
plt.plot(house_sizes, predicted_prices, color='red', label='Predicted Line')
plt.xlabel('House Size (sqft)')
plt.ylabel('Price ($)')
plt.title('House Price Prediction using Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```
