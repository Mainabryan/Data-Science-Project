```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression # Import LinearRegression from sklearn.linear_model

# column sizes in soft,number of bedrooms
X = np.array([
    [1000, 2],
    [1500, 3],
    [2000, 3],
    [2500, 4],
    [3000, 4]
])
# Target: House Prices
y = np.array([200000, 300000, 400000, 500000, 600000])

# Train the model
model = LinearRegression()
model.fit(X, y)
```

# predict

```python
new_house = np.array([[2300,3]])
predicted_price = model.predict(new_house)
print(f"Predicted price for a 2300 sqft house with 3 bedrooms: ${predicted_price[0]:.2f}")
```

# plotting

```python
plt.scatter(X[:, 0], y, color='blue', label='Actual Prices')
plt.plot(X[:, 0], model.predict(X), color='red', label='Predicted Prices')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($)')
plt.title('House Price vs Size')
plt.grid(True)
plt.legend()
plt.show()
```
