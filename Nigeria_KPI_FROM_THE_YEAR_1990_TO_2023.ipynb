#5. Perform Data Transformation on the columns

```python
#lets transform  the inflation rate ,Unemployment Rates and Government Debt Rate
df['Inflation rate'] = df['Inflation rate']*100
df['Unemployment Rates'] = df['Unemployment Rates']*100
df['Government Debt Rate'] = df['Government Debt Rate']*100

df.head()
```
