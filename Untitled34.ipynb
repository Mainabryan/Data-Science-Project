```python
# Define a class to represent a Book

# Initialize the book with title, author, ISBN, and publisher
# Also, set the availability status of the book (default is False)

# Function to check if the book is available
# If the book is available, return a message stating that it is available
# Otherwise, return a message stating that it is not available

# Function to return book information (title, author, ISBN, and publisher)

# Function to mark the book as returned
# If the book was not available, update its status to available
# If the book was already available, return a message indicating that

# Create a list of book objects with predefined book details

# Define a function to search for a book based on user input

# Convert the search query to lowercase for case-insensitive search

# Loop through each book in the list

# Check if the query matches the title, author, ISBN, or publisher

# If a matching book is found:
# - Display book found message
# - Display book details (title, author, ISBN, publisher)
# - Display book availability status

# If no book matches, display book not found message

# Ask the user to enter a book title, author, ISBN, or publisher to search

# Call the search function with the user input
class Book:
    def __init__(self, title, author, isbn, publisher):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.publisher = publisher
        self.available = False  # Changed from list to boolean

    def available_book(self):
        if self.available:
            return f"{self.title} by {self.author} is available."
        else:
            return f"{self.title} by {self.author} is not available."

    def get_book_info(self):
        return f"{self.title} by {self.author} has ISBN {self.isbn} and is published by {self.publisher}."

    def return_book(self):
        if not self.available:
            self.available = True
            return f"{self.title} by {self.author} has been returned."
        else:
            return f"{self.title} by {self.author} is already available."

# Create book instances
books = [
    Book("The Alchemist", "Paulo Coelho", "978-0-06-226814-1", "HarperCollins"),
    Book("The Lord of the Rings", "J.R.R. Tolkien", "978-0-618-64015-1", "Houghton Mifflin Harcourt"),
    Book("The Hunger Games", "Suzanne Collins", "978-0-439-02348-1", "Scholastic")
]

# Function to search for books
def search_book(query):
    query = query.lower()
    for book in books:
        if (query in book.title.lower() or query in book.author.lower() or
            query in book.isbn.lower() or query in book.publisher.lower()):
            print("\nBook Found!")
            print(book.get_book_info())
            print(book.available_book())
            return
    print("\nBook not found.")

# Get user input
book_input = input("Enter the book title, author, ISBN, or publisher to search: ")
search_book(book_input)
```

```python
# Define a class named Movie
# Create an initializer (__init__) that takes title, director, genre, release_year, and availability

# Create a method to display movie information
# Create a method to check if the movie is available
# Create a method to mark the movie as returned (available)

# Create instances of at least three movies

# Store the movie instances in a list

# Define a function to search for a movie by title, director, genre, or release year
# Convert the userâ€™s input to lowercase for case-insensitive search
# Loop through the movie list to check if the query matches any movie attributes
# If a match is found, display movie information and availability
# If no match is found, print "Movie not found."

# Ask the user to enter a search query
# Call the search function with the user input
class movie:
    def __init__(self, title, director, genre, release_year, availability):
        self.title = title
        self.director = director
        self.genre = genre
        self.release_year = release_year
        self.availability = True

    # Added proper indentation to the following methods:
    def display_movie_info(self):
        return f"{self.title} by {self.director} is a {self.genre} movie released in {self.release_year}"

    def movie_available(self):
        if self.availability: # Corrected variable name here (self_availability to self.availability)
            return f"{self.title} by {self.director} is available"
        else:
            return f"{self.title} by {self.director} is not available"

    def return_movie(self):
        if not self.availability:
            self.availability = True
            return f"{self.title} by {self.director} has been returned."
        else:
            return f"{self.title} by {self.director} is already available."

# Create movie instances and store them in a list
movies = [
    movie("The Shawshank Redemption", "Frank Darabont", "Drama", 1994, True),
    movie("The Godfather", "Francis Ford Coppola", "Crime", 1972, True),
    movie("The Dark Knight", "Christopher Nolan", "Action", 2008, True)
]

# Function to search for a movie
def search_movie(query):
    query = query.lower()
    for movie in movies:
        if (query in movie.title.lower() or query in movie.director.lower() or
            query in movie.genre.lower() or query in str(movie.release_year).lower()):
            print("\nMovie available") # Changed message to "Movie available"
            print(movie.display_movie_info())
            print(movie.movie_available())
            return
    print("\nMovie not found.")

# Get user input and call the search function
movie_input = input("Enter the movie title, director, genre, or release year to search: ")
search_movie(movie_input)
```
