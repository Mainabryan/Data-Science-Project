```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # Corrected the module name from 'metplotlib' to 'matplotlib'
from sklearn.linear_model import LinearRegression # Corrected the module and class name for Linear Regression
```

```python
# Step 1: Data
distance_km = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
fare_ksh = np.array([100, 150, 200, 250, 300, 350, 400])  # in Kenyan Shillings (KSh)
model = LinearRegression()
model.fit(distance_km,fare_ksh)
# Reshape the input for prediction to be a 2D array
predicted_prices = model.predict(distance_km)
plt.scatter(distance_km,fare_ksh, color = 'blue',label = 'Actual prices')
plt.plot(distance_km, predicted_prices, color ='red',label = 'Predicted prices') # Plot distance_km on the x-axis
plt.xlabel('Distance in km')
plt.ylabel('Fare in KSh')
plt.title('Distance vs Fare in KSh')
plt.legend()
plt.show()
```
