# import the necessay ibraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

# perform the comparison of th data

```python
distance_km = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
fare_ksh = np.array([120, 140, 230, 270, 310, 355, 415])
```

# Train the model

```python
model = LinearRegression()
model.fit(distance_km, fare_ksh)

```

# Input the data

```python
user_input = float(input('Enter the distance in km:'))
predicted = model.predict (np.array([[user_input]]))
print(f'predicted fare for {user_input} km is {predicted[0]:.2f}ksh')
```

```python
predicted_fares = model.predict(distance_km)
plt.scatter(distance_km, fare_ksh)
plt.plot(distance_km, predicted_fares, color='red')
plt.xlabel('Distance (km)')
plt.ylabel('Fare (ksh)')
plt.title('Distance vs Fare')
plt.show()
plt.legend()
plt.grid(True)
plt.show
```
